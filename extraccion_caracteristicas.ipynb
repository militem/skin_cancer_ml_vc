```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from os import listdir

def momentos(_dir, _y, _n):
    direc = _dir
    names = listdir(direc)
    xp = [0, 64, 128, 192, 255]
    fp = [0, 16, 128, 240, 255]
    x = np.arange(256)
    table = np.interp(x, xp, fp).astype('uint8')
    train_images = []
    for i in range(0,_n):
        try:
            filename = direc + names[i]
            image = plt.imread(filename)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            blur = cv2.GaussianBlur(gray,(5,5),0)
            img2 = cv2.LUT(blur, table)
            M = cv2.moments(img2)
            M = list(M.values())
            M.append(_y)
            train_images.append(M)
        except:
            pass
    return train_images

        
data_b = momentos("C:\\Users\\Alex\\Downloads\\melanoma_cancer_dataset\\dataset\\benign\\", 0, 500)
data_m = momentos("C:\\Users\\Alex\\Downloads\\melanoma_cancer_dataset\\dataset\\malignant\\", 1, 500)
data = data_b + data_m
print(data)
```

```python
from skimage import data
from skimage import transform
from skimage.color import rgb2gray
from skimage.feature import match_descriptors, plot_matches, SIFT

def sift_csv(_dir, _y, _n):
    descriptor_extractor = SIFT()
    direc = _dir
    names = listdir(direc)
    n = (len(listdir(direc)))
    xp = [0, 64, 128, 192, 255]
    fp = [0, 16, 128, 240, 255]
    x = np.arange(256)
    table = np.interp(x, xp, fp).astype('uint8')
    images = []
    for i in range(0,_n):
        try:
            filename = direc + names[i]
            image = plt.imread(filename)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            blur = cv2.GaussianBlur(gray,(5,5),0)
            img2 = cv2.LUT(blur, table)
            descriptor_extractor.detect_and_extract(img2)
            descriptors = descriptor_extractor.descriptors
            aux = list(descriptors[0][0:24])
            aux.append(_y)
            images.append(aux)
        except:
            print("Problemas con la imagen: " + names[i])
            pass
    return images

data_sift_b = sift_csv("C:\\Users\\Alex\\Downloads\\melanoma_cancer_dataset\\dataset\\benign\\", 0, 500)
data_sift_m = sift_csv("C:\\Users\\Alex\\Downloads\\melanoma_cancer_dataset\\dataset\\malignant\\", 1, 500)
```

```python
data_sift = data_sift_b + data_sift_m
print(data_sift)
```

```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure

def hog_ft(_dir, _y, _n):
    direc = _dir
    names = listdir(direc)
    n = (len(listdir(direc)))
    xp = [0, 64, 128, 192, 255]
    fp = [0, 16, 128, 240, 255]
    x = np.arange(256)
    table = np.interp(x, xp, fp).astype('uint8')
    images = []
    for i in range(0,_n):
        try:
            filename = direc + names[i]
            image = plt.imread(filename)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            blur = cv2.GaussianBlur(gray,(5,5),0)
            img2 = cv2.LUT(blur, table)
            resized_img = resize(img2, (128*4, 64*4))
            fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                cells_per_block=(2, 2), visualize=True, multichannel=False)
            aux = list(fd[0:24])
            aux.append(_y)
            images.append(aux)
        except:
            print("Problemas con la imagen: " + names[i])
            pass
    return images

data_hog_b = hog_ft("C:\\Users\\Alex\\Downloads\\melanoma_cancer_dataset\\dataset\\benign\\", 0, 500)
data_hog_m = hog_ft("C:\\Users\\Alex\\Downloads\\melanoma_cancer_dataset\\dataset\\malignant\\", 1, 500)

filename = "C:\\Users\\Alex\\Downloads\\melanoma_cancer_dataset\\dataset\\benign\\melanoma_77.jpg"
img = plt.imread(filename)
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.axis("off")
plt.imshow(img,'gray')
print(img.shape)
```
